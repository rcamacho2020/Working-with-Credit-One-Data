# Outliers

```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(credit))
print(z)
```

```python
threshold = 3
print(np.where(z > 3))
```

```python
print(z[6][12])
```


The interquartile range is calculated: IQR = Q3 – Q1.
A commonly used rule says that a data point is an outlier if it is more than 1.5 * IQR above the third quartile, or below the first quartile. 
Said differently, low outliers are below Q1 − 1.5 * IQR and high outliers are above Q3 + 1.5 * IQR


```python
Q1 = credit.quantile(0.25)
Q3 = credit.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
lower = credit < (Q1 - 1.5 * IQR)
print(lower)
```

```python
upper = credit > (Q3 + 1.5 * IQR)
print(upper)
```

```python
Credit_o = credit[(z < 3).all(axis=1)]
print (Credit_o)
```

```python
Credit_out = lower | upper
print (Credit_out)
```

```python
Credit_out = Credit_o[~((Credit_o < (Q1 - 1.5 * IQR)) |(Credit_o > (Q3 + 1.5 * IQR))).any(axis=1)]
Credit_out.shape
```
